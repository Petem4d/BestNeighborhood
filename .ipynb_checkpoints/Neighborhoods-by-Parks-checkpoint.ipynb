```python
import pandas as pd

##Pittsburgh_City-parks.cvs is from https://data.wprdc.org/dataset/parks
##Imports Pittsburgh_City-Parks.cvs as a dataframe parks
## Only grabs the Collums Name & Neighborhood, as they are all that will be relevant to my usage
parks = pd.read_csv("Pittsburgh_City-Parks.csv", usecols = ["name","type","neighborhood"])

parks
```

```python

neighToParks = {}

##Stores each neighborhood paired with all the parks in that neighborhood
for index, row in parks.iterrows():
    if row["type"] == "Park":
        key = row["neighborhood"]
        if not(key in neighToParks.keys()):
            neighToParks[key] = []
        neighToParks[key].append(row["name"])
sorted_neighToParks = {}

##Storts the negihborhoods in Alphabetic order
for i in sorted (neighToParks.keys()) :
    sorted_neighToParks[i] = neighToParks[i]
```

```python
neigh = pd.DataFrame({"neighborhood": [], "parks": [], "noParks": []})
i = 0
for key in sorted_neighToParks.keys():
    neigh.loc[i]=[key,sorted_neighToParks[key],int(len(sorted_neighToParks[key]))]
    i += 1
neigh
```
